# Frequentism and Bayesianism V: Model Selection

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import seaborn; seaborn.set()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

def PolynomialRegression(degree, *args, **kwargs):
    return make_pipeline(PolynomialFeatures(degree),
                         LinearRegression(*args, **kwargs))
```

```python
def polyfit(degree=2):
    rng = np.random.RandomState(0)
    x = rng.rand(50)
    dy = 0.1
    y = x ** 2 + dy * rng.randn(x.size)
    plt.plot(x, y, 'ok')

    xfit = np.linspace(0, 1, 1000)
    poly = PolynomialRegression(degree)
    poly.fit(x[:, None], y)
    yfit = poly.predict(xfit[:, None])
    
    logL = -0.5 * np.sum((poly.predict(x[:, None]) - y) ** 2)

    plt.plot(xfit, yfit, '-k')
    plt.text(0.05, 1, r"$\log L_0 = {0:.3f}$".format(logL),
             size=18)
    plt.xlim(0, 1)
    plt.ylim(-0.1, 1.1)
```

```python
from IPython.html.widgets import interact
interact(polyfit, degree=[0, 50]);
```
